```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
samplerate = 44100

def apply_sin(phi):
    return np.sin(phi)

def get_signal(signal_func, frequency=440.0, amplitude=1, phase=0.0, samplerate=samplerate, duration=1):
    base = 2 * np.pi * frequency / samplerate
    phi = phase
    values = []    
    for i in range(samplerate * duration):
        phi += base

        signal_value = signal_func(phi)   
        
        values.append( amplitude * signal_value)
    return np.array(values) 

def get_poly(signals, samplerate=samplerate, duration=1):
    poly = np.zeros(samplerate * duration)
    for signal in signals:
        poly += signal
    return poly   
```

```python
sin = get_signal(apply_sin, frequency=550)

plt.plot(np.arange(0,2048), sin[:2048])
```

```python
f_sin = np.fft.rfft(sin)
```

```python
plt.plot(np.arange(0,2048),f_sin[:2048])
```
